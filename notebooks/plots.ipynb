# Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

# Data

```python
x = np.arange(-10,11,1)
y = x**2
plt.plot(x,y,color='orange',linewidth=3.0)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('parabel.png', transparent=True)
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='orange',linewidth=3.0)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('normal_distribution.png', transparent=True)
```
