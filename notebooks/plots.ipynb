```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

# Plots

```python
x = np.arange(-10,11,1)
y = x**2
plt.plot(x,y,color='orange',linewidth=3.0)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('parabel.png', transparent=True)
```

```python
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='orange',linewidth=3.0)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('normal_distribution.png', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(5, 2))
x = np.arange(-10,11,1)
y = 1/(1 + np.exp(-x))
plt.plot(x,y,color='orange',linewidth=3.0)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.savefig('logistic_function.png', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(-10.5,10.5,1)
y = 1/(1 + np.exp(-x))
plt.plot(x,y,color='orange',linewidth=3.0)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines[['right', 'top']].set_visible(False)
ax.hlines(0,-10.5,10.5, linestyle='dashed',color="grey")
ax.hlines(1,-10.5,10.5, linestyle='dashed',color="grey")
fig.text(0.6, .6, "$f(z)= \\frac{1}{1+e^{-z}}$", fontsize=14)
plt.savefig('logistic_function_values.png', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(-10.5,10.5,1)
a = 1
b = 0
y = 1/(1 + np.exp(-(a*x+b)))
plt.plot(x,y,color='orange',linewidth=3.0)
a = 0.5
b = 0
y = 1/(1 + np.exp(-(a*x+b)))
plt.plot(x,y,color='#5B2C6F',linewidth=1.0, linestyle='dashed', label='a=0.5, b=0')
a = 1.5
b = 0
y = 1/(1 + np.exp(-(a*x+b)))
plt.plot(x,y,color='#2980B9',linewidth=1.0, linestyle='dashed', label='a=1.5, b=0')
a = 0.5
b = 1
y = 1/(1 + np.exp(-(a*x+b)))
plt.plot(x,y,color='#138D75',linewidth=1.0, linestyle='dashed', label='a=0.5, b=1')
plt.legend()


# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines[['right', 'top']].set_visible(False)
ax.hlines(0,-10.5,10.5, linestyle='dashed',color="grey")
ax.hlines(1,-10.5,10.5, linestyle='dashed',color="grey")
fig.text(0.6, .6, "$f(x)= \\frac{1}{1+e^{-a\cdot x - b}}$", fontsize=14)
plt.savefig('logistic_function_different_parameters.png', transparent=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
x = np.arange(0,1.1,.005)
y = -np.log(x)
ax.plot(x,y,color='orange',linewidth=3.0)
ax.spines[['right', 'top']].set_visible(False)
ax.set_xlabel('$\hat{y}$', fontsize=14)
ax.set_ylabel('$-\log{\hat{y}}$', fontsize=14)
plt.savefig('logloss1.png', transparent=True)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8,3))
x = np.arange(0,1.01,0.01)
y1 = 1 - (x*x) - (1-x)*(1-x) 
y2 = - (x*np.log2(x)) - (1-x)*np.log2(1-x) 

ax[0].plot(x,y1,color='orange',linewidth=3.0)
ax[0].spines[['right', 'top']].set_visible(False)
ax[0].set_title('Gini Impurity')
ax[1].plot(x,y2,color='orange',linewidth=3.0)
ax[1].spines[['right', 'top']].set_visible(False)
ax[1].set_title('Entropy');
plt.savefig('impurity_entropy.png', transparent=True)
```

**Bias / Variance**

```python
np.random.seed(42)
x = np.arange(0,8,0.3)
noise = np.random.normal(0,0.1,x.shape[0])
y = np.sin(x)+noise
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,3))
y1 = 0.1*x-0.5
ax[0].scatter(x, y, color='lightgrey')
ax[0].plot(x,y1, color='orange',linewidth=3.0)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_title('High Bias / Low Variance', fontsize=18)
ax[0].set_xlabel('Underfitting', fontsize=18); 
y1 = np.sin(x)
ax[1].scatter(x, y, color='lightgrey')
ax[1].plot(x,y1, color='orange',linewidth=3.0)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('Optimal Bias', fontsize=18)
ax[1].set_xlabel('Optimal fit', fontsize=18); 
y1 = np.sin(x)+noise
ax[2].scatter(x, y, color='lightgrey')
ax[2].plot(x,y1, color='orange',linewidth=3.0)
ax[2].set_xticks([])
ax[2].set_yticks([])
ax[2].set_xlabel('Overfitting', fontsize=18)
ax[2].set_title('Low Bias / High Variance', fontsize=18); 
plt.savefig('bias.png', transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 2, figsize=(12,16))
np.random.seed(42)
x = np.arange(0,30,1)
noise = np.random.normal(0, 500,x.shape[0])
a0 = 20
a1 = 0.3
a2 = -25
a3 = 1.1
y = a0 + a1*x + a2*x**2 + a3*x*x*x
# training data
y1 = y + noise
# models
# 1. linear
model1 = np.poly1d(np.polyfit(x, y1, 1))
# 2. quadratic
model2 = np.poly1d(np.polyfit(x, y1, 2))
# 3. cubic
model3 = np.poly1d(np.polyfit(x, y1, 3))
# 4. degree 10
model4 = np.poly1d(np.polyfit(x, y1, 10))
# 5. degree 15
model5 = np.poly1d(np.polyfit(x, y1, 15))
# 6. degree 20
model6 = np.poly1d(np.polyfit(x, y1, 20))


ax[0,0].scatter(x, y1, color='lightgrey', label="training data")
ax[0,0].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[0,0].plot(x, model1(x), color='darkcyan',linewidth=3.0, label="linear model")
ax[0,0].legend();

ax[0,1].scatter(x, y1, color='lightgrey', label="training data")
ax[0,1].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[0,1].plot(x, model2(x), color='darkcyan',linewidth=3.0, label="quadratic model")
ax[0,1].legend();

ax[1,0].scatter(x, y1, color='lightgrey', label="training data")
ax[1,0].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[1,0].plot(x, model3(x), color='darkcyan',linewidth=3.0, label="cubic model")
ax[1,0].legend();

ax[1,1].scatter(x, y1, color='lightgrey', label="training data")
ax[1,1].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[1,1].plot(x, model4(x), color='darkcyan',linewidth=3.0, label="model degree 10")
ax[1,1].legend();

ax[2,0].scatter(x, y1, color='lightgrey', label="training data")
ax[2,0].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[2,0].plot(x, model5(x), color='darkcyan',linewidth=3.0, label="model degree 15")
ax[2,0].legend();

ax[2,1].scatter(x, y1, color='lightgrey', label="training data")
ax[2,1].plot(x, y, color='orange',linewidth=3.0, label="true model")
ax[2,1].plot(x, model6(x), color='darkcyan',linewidth=3.0, label="model degree 20")
ax[2,1].legend();

plt.savefig('model_complexity.png', transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 

np.random.seed(42)
x = np.arange(0,30,0.5)
a0 = 20
a1 = 0.3
a2 = -25
a3 = 1.1
y = a0 + a1*x + a2*x**2 + a3*x*x*x
noise = np.random.normal(0, 500,x.shape[0])
# training data
y1 = y + noise

x_train1, x_test1, y_train1, y_test1 = train_test_split(x, y1, test_size=0.33, random_state=42)
x_train2, x_test2, y_train2, y_test2 = train_test_split(x, y1, test_size=0.33, random_state=1)
x_train3, x_test3, y_train3, y_test3 = train_test_split(x, y1, test_size=0.33, random_state=2)
x_train4, x_test4, y_train4, y_test4 = train_test_split(x, y1, test_size=0.33, random_state=3)
x_train5, x_test5, y_train5, y_test5 = train_test_split(x, y1, test_size=0.33, random_state=4)


fig, ax = plt.subplots(3, 2, figsize=(12,16))

# models
# 1. linear
model1_1 = np.poly1d(np.polyfit(x_train1, y_train1, 1))
model1_2 = np.poly1d(np.polyfit(x_train2, y_train2, 1))
model1_3 = np.poly1d(np.polyfit(x_train3, y_train3, 1))
model1_4 = np.poly1d(np.polyfit(x_train4, y_train4, 1))
model1_5 = np.poly1d(np.polyfit(x_train5, y_train5, 1))
# 2. quadratic
model2_1 = np.poly1d(np.polyfit(x_train1, y_train1, 2))
model2_2 = np.poly1d(np.polyfit(x_train2, y_train2, 2))
model2_3 = np.poly1d(np.polyfit(x_train3, y_train3, 2))
model2_4 = np.poly1d(np.polyfit(x_train4, y_train4, 2))
model2_5 = np.poly1d(np.polyfit(x_train5, y_train5, 2))
# 3. cubic
model3_1 = np.poly1d(np.polyfit(x_train1, y_train1, 3))
model3_2 = np.poly1d(np.polyfit(x_train2, y_train2, 3))
model3_3 = np.poly1d(np.polyfit(x_train3, y_train3, 3))
model3_4 = np.poly1d(np.polyfit(x_train4, y_train4, 3))
model3_5 = np.poly1d(np.polyfit(x_train5, y_train5, 3))
# 4. degree 10
model4_1 = np.poly1d(np.polyfit(x_train1, y_train1, 10))
model4_2 = np.poly1d(np.polyfit(x_train2, y_train2, 10))
model4_3 = np.poly1d(np.polyfit(x_train3, y_train3, 10))
model4_4 = np.poly1d(np.polyfit(x_train4, y_train4, 10))
model4_5 = np.poly1d(np.polyfit(x_train5, y_train5, 10))
# 5. degree 15
model5_1 = np.poly1d(np.polyfit(x_train1, y_train1, 15))
model5_2 = np.poly1d(np.polyfit(x_train2, y_train2, 15))
model5_3 = np.poly1d(np.polyfit(x_train3, y_train3, 15))
model5_4 = np.poly1d(np.polyfit(x_train4, y_train4, 15))
model5_5 = np.poly1d(np.polyfit(x_train5, y_train5, 15))
# 6. degree 20
model6_1 = np.poly1d(np.polyfit(x_train1, y_train1, 20))
model6_2 = np.poly1d(np.polyfit(x_train2, y_train2, 20))
model6_3 = np.poly1d(np.polyfit(x_train3, y_train3, 20))
model6_4 = np.poly1d(np.polyfit(x_train4, y_train4, 20))
model6_5 = np.poly1d(np.polyfit(x_train5, y_train5, 20))

ax[0,0].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[0,0].plot(x, y, color='orange', label="true model", linewidth=1)
ax[0,0].scatter(x_test1, model1_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[0,0].scatter(x_test2, model1_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[0,0].scatter(x_test3, model1_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[0,0].scatter(x_test4, model1_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[0,0].scatter(x_test5, model1_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[0,0].set_title('linear model')
ax[0,0].set_ylim(-2500, 6500)
ax[0,0].legend();

ax[0,1].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[0,1].plot(x, y, color='orange',linewidth=1, label="true model")
ax[0,1].scatter(x_test1, model2_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[0,1].scatter(x_test2, model2_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[0,1].scatter(x_test3, model2_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[0,1].scatter(x_test4, model2_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[0,1].scatter(x_test5, model2_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[0,1].set_title('quadratic model')
ax[0,1].set_ylim(-2500, 6500)
ax[0,1].legend();

ax[1,0].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[1,0].plot(x, y, color='orange',linewidth=1.0, label="true model")
ax[1,0].scatter(x_test1, model3_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[1,0].scatter(x_test2, model3_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[1,0].scatter(x_test3, model3_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[1,0].scatter(x_test4, model3_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[1,0].scatter(x_test5, model3_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[1,0].set_title('cubic model')
ax[1,0].set_ylim(-2500, 6500)
ax[1,0].legend();

ax[1,1].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[1,1].plot(x, y, color='orange',linewidth=1.0, label="true model")
ax[1,1].scatter(x_test1, model4_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[1,1].scatter(x_test2, model4_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[1,1].scatter(x_test3, model4_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[1,1].scatter(x_test4, model4_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[1,1].scatter(x_test5, model4_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[1,1].set_title('model degree 10')
ax[1,1].set_ylim(-2500, 6500)
ax[1,1].legend();

ax[2,0].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[2,0].plot(x, y, color='orange',linewidth=1.0, label="true model")
ax[2,0].set_title('model degree 15')
ax[2,0].scatter(x_test1, model5_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[2,0].scatter(x_test2, model5_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[2,0].scatter(x_test3, model5_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[2,0].scatter(x_test4, model5_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[2,0].scatter(x_test5, model5_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[2,0].set_ylim(-2500, 6500)
ax[2,0].legend();

ax[2,1].scatter(x, y1, color='lightgrey', label="training data", s=3)
ax[2,1].plot(x, y, color='orange',linewidth=1.0, label="true model")
ax[2,1].scatter(x_test1, model6_1(x_test1), color='darkcyan',s=3.0, label="split 1", alpha=0.5)
ax[2,1].scatter(x_test2, model6_2(x_test2), color='blue', s=3.0, label="split 2", alpha=0.5)
ax[2,1].scatter(x_test3, model6_3(x_test3), color='green', s=3.0, label="split 3", alpha=0.5)
ax[2,1].scatter(x_test4, model6_4(x_test4), color='darkgreen', s=3.0, label="split 4", alpha=0.5)
ax[2,1].scatter(x_test5, model6_5(x_test5), color='cyan', s=3.0, label="split 5", alpha=0.5)
ax[2,1].set_title('model degree 20')
ax[2,1].set_ylim(-2500, 6500)
ax[2,1].legend();

plt.savefig('model_complexity.png', transparent=True)
```

```python
# plot influence ("amount of say")
fig, ax = plt.subplots(1, 1, figsize=(6,6))

err = np.arange(0,1.01,0.01)
alpha = 1/2 *np.log((1-err)/err)
ax.plot(err, alpha, color="orange", linewidth=3)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines[['right', 'top']].set_visible(False)
ax.set_xlabel("                                                        Total Error", fontsize=14)
ax.set_ylabel("                                                        alpha", fontsize=14) 
plt.savefig('alpha_adaboost.png', transparent=True);
```
