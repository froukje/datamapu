```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x = np.linspace(-1.0,1.0,100)
noise = np.random.normal(0,0.5,100)
y = x + noise
```

```python
x
```

```python
data = {}
data["feature 1"] = x
data["feature 2"] = y
df = pd.DataFrame(data)
df
```

```python
plt.scatter(x=df["feature 1"], y=df["feature 2"])
plt.axvline(x=0, color='grey', linestyle='--', label='y-axis')
plt.axhline(y=0, color='grey', linestyle='--', label='x-axis')
plt.xlim(-2,2)
plt.ylim(-2.5,2.5)

vector1 = np.array([0.53085838, 0.84746055])
slope1 = vector1[1] / vector1[0]  # Slope = b/a
x_vals = np.array(plt.gca().get_xlim())  
y_vals1 = slope1 * x_vals1  

vector2 = np.array([0.53085838, -0.84746055])
slope2 = vector2[1] / vector2[0]  # Slope = b/a  
y_vals2 = slope2 * x_vals 

plt.plot(x_vals, y_vals1, color='orange', linestyle='--')
plt.plot(x_vals, y_vals2, color='orange', linestyle='--')
plt.savefig('pca_example1.png', transparent=True);
```

```python
pca = PCA(n_components=2)
pca.fit(df)
```

```python
print(pca.explained_variance_ratio_)
```

```python
print(pca.singular_values_)
```

```python
print(pca.components_)
```

```python
components = pca.fit_transform(df)
```

```python
plt.scatter(x=components[:,0], y=components[:,1])
plt.xlim(-2,2)
plt.ylim(-2.5,2.5)
plt.axvline(x=0, color='orange', linestyle='--', label='y-axis')
plt.axhline(y=0, color='orange', linestyle='--', label='x-axis')
plt.savefig('pca_example2.png', transparent=True);
```
